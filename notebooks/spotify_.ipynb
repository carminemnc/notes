```python
df = pd.read_csv('../data/podcast_dataset.csv').head(5000)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Create 3D scatter plot
scatter = ax.scatter(df['Host_Popularity_percentage'], 
                    df['Guest_Popularity_percentage'],
                    df['Listening_Time_minutes'],
                    c=df['Listening_Time_minutes'],
                    cmap='cool',
                    alpha=0.2,
                    s=70)

# Remove absolutely everything
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_zticklabels([])
ax.axis('off')

# Make all panes completely invisible
ax.xaxis.pane.fill = False
ax.yaxis.pane.fill = False
ax.zaxis.pane.fill = False
ax.xaxis.pane.set_edgecolor('none')
ax.yaxis.pane.set_edgecolor('none')
ax.zaxis.pane.set_edgecolor('none')
ax.grid(False)
ax._axis3don = False  # This removes the final edges

# Adjust viewing angle
ax.view_init(elev=20, azim=45)

# Add stronger glow effect
mplcyberpunk.make_scatter_glow()

plt.tight_layout()
plt.show()
```
